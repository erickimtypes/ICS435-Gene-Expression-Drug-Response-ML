```python
import pandas as pd
from IPython.display import display
```

```python
expr = pd.read_csv('../data/raw/OmicsExpressionTPMLogp1HumanProteinCodingGenes.csv')
expr_meta = pd.read_csv('../data/raw/Model.csv')
drug = pd.read_csv('../data/raw/Repurposing_Public_24Q2_Extended_Primary_Data_Matrix.csv')
drug_meta = pd.read_csv('../data/raw/Repurposing_Public_24Q2_Extended_Primary_Compound_List.csv')

datasets = {
    'Expression': expr,
    'Expression Metadata': expr_meta,
    'Drug response': drug,
    'Drug Response Metadata': drug_meta,
}

for name, df in datasets.items():
    print(f"{name}: {df.shape[0]:,} rows x {df.shape[1]:,} columns")
    print(f"Columns: {df.columns}")
    display(df.head())
    print("\n")
```

```python
print(f"Number of unique ModelIDs in Expression dataset: {expr['ModelID'].nunique()}")
print(f"Number of unique ModelIDs in Expression Metadata: {expr_meta['ModelID'].nunique()}")
print(f"Number of columns in PRISM dataset (cell lines): {len(drug.columns)}")  

common = set(expr['ModelID']).intersection(set(drug.columns))
print(f"\nNumber of overlapping cell lines: {len(common)}")
```

```python
drug = drug.rename(columns={'Unnamed: 0': 'IDs'})

common_ids = list(common)

expr_small = expr[expr['ModelID'].isin(common_ids)]

drug_small = drug[['IDs'] + common_ids]

prism_long = drug_small.melt(
    id_vars='IDs',
    var_name='ModelID',
    value_name='Response'
)

prism_long = prism_long.merge(
    drug_meta,
    on='IDs',
    how='left'
)

print(prism_long.shape)
print(prism_long[['IDs', 'Drug.Name', 'ModelID', 'Response']].head())
print(prism_long['Drug.Name'].value_counts().head(20))
```

```python
import os

def build_drug_dataset(drug_name, prism_long, expr_df, save=True, save_dir="../data/processed"):
    """
    Builds a merged gene expression + drug response dataset for a single drug.
    
    Parameters:
    - drug_name (str): name of the drug (case-insensitive)
    - prism_long (DataFrame): long-format PRISM data
    - expr_df (DataFrame): filtered gene expression data (expr_small)
    - save (bool): whether to save the dataset
    - save_dir (str): directory to save file
    
    Returns:
    - final_df (DataFrame)
    """
    
    # Filter for selected drug
    one_drug = prism_long[
        prism_long['Drug.Name'].str.lower() == drug_name.lower()
    ].copy()
    
    # Drop missing response values
    one_drug = one_drug.dropna(subset=['Response'])
    
    # Merge with gene expression
    final_df = one_drug.merge(
        expr_df,
        on='ModelID',
        how='inner'
    )
    
    print(f"\nDrug: {drug_name}")
    print(f"Shape: {final_df.shape}")
    
    # Optional save
    if save:
        os.makedirs(save_dir, exist_ok=True)
        
        # Clean filename (remove spaces, lowercase)
        safe_name = drug_name.lower().replace(" ", "_")
        filepath = f"{save_dir}/{safe_name}_expression_response.csv"
        
        final_df.to_csv(filepath, index=False)
        print(f"Saved to: {filepath}")
    
    return final_df
```

### Why I am using one drug (for now)

My overall goal is to build a model that takes gene expression data and a drug as input, and predicts how a cancer cell will respond.

Right now, I am starting with **one drug only**. This means my current model learns:

**gene expression → response to one specific drug**

I am doing this because it helps me first learn the core relationship between gene expression and drug response. If the model cannot learn this for one drug, it will not work for multiple drugs.

Later, I can extend this by:

* Training models for multiple drugs, or
* Building a more advanced model that takes both gene expression and drug as input.

So this step is building the **foundation** for the final goal.
